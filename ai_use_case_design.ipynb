```python
high_risk = df[
    (df["Contract"]=="Month-to-month") &
    (df["MonthlyCharges"]>80)
]

print("High Risk Customers")

high_risk.head()
```

```python
print(len(high_risk))
```

```python
def recommend(contract,charges):

    if contract=="Month-to-month" and charges>80:
        return "Offer 20% Discount"

    elif contract=="Month-to-month":
        return "Upgrade to Annual Plan"

    elif charges>80:
        return "Premium Loyalty Offer"

    else:
        return "No Action Needed"
```

```python
df["AI_Recommendation"] = df.apply(

lambda x: recommend(

x["Contract"],

x["MonthlyCharges"]

),

axis=1

)

df.head()
```

```python
df["AI_Recommendation"].value_counts()
```

```python
plt.figure(figsize=(10,5))

sns.countplot(
    data=df,
    x="AI_Recommendation"
)

plt.xticks(rotation=20)

plt.show()
```
